```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("Employee.csv")
```

```python
data
```

```python
education_counts = data['Education'].value_counts()
plt.figure(figsize=(10, 6))
education_counts.plot(kind='bar', color='blue')
plt.title('Distribution of Education Levels')
plt.xlabel('Education Level')
plt.ylabel('Number of Employees')
plt.xticks(rotation=0)
plt.show()

payment_tier_counts = data['PaymentTier'].value_counts()
plt.figure(figsize=(10, 6))
payment_tier_counts.plot(kind='bar', color='violet')
plt.title('Distribution of Payment Tiers')
plt.xlabel('Payment Tier')
plt.ylabel('Number of Employees')
plt.xticks(rotation=0)
plt.show()

age_gender_counts = data.groupby(['Age', 'Gender']).size().unstack()
age_gender_counts.plot(kind='bar', figsize=(12, 8), color=['red', 'yellow'])
plt.title('Distribution of Age and Gender')
plt.xlabel('Age')
plt.ylabel('Number of Employees')
plt.xticks(rotation=0)
plt.legend(title='Gender')
plt.show()




# Assuming 'data' is your DataFrame and 'ExperienceInCurrentDomain' is the column of interest
experience_counts = data[data['ExperienceInCurrentDomain']<=5]['ExperienceInCurrentDomain'].value_counts()



plt.figure(figsize=(10, 6))
plt.pie(experience_counts, labels=experience_counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors,
        wedgeprops=dict(width=0.3, edgecolor='w'), pctdistance=0.85 )
plt.title('Distribution of Experience in Current Domain')

# Draw a circle at the center of pie to make it look like a donut
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')
plt.show()

```
