```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("Employee.csv")
```

```python
data
```

```python
education_counts = data['Education'].value_counts()
plt.figure(figsize=(10, 6))
education_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Education Levels')
plt.xlabel('Education Level')
plt.ylabel('Number of Employees')
plt.xticks(rotation=0)
plt.show()

payment_tier_counts = data['PaymentTier'].value_counts()
plt.figure(figsize=(10, 6))
payment_tier_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Payment Tiers')
plt.xlabel('Payment Tier')
plt.ylabel('Number of Employees')
plt.xticks(rotation=0)
plt.show()

age_gender_counts = data.groupby(['Age', 'Gender']).size().unstack()
age_gender_counts.plot(kind='bar', figsize=(12, 8), color=['skyblue', 'lightcoral'])
plt.title('Distribution of Age and Gender')
plt.xlabel('Age')
plt.ylabel('Number of Employees')
plt.xticks(rotation=0)
plt.legend(title='Gender')
plt.show()

plt.figure(figsize=(10, 6))
plt.hist(data['ExperienceInCurrentDomain'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Experience in Current Domain')
plt.xlabel('Years of Experience')
plt.ylabel('Number of Employees')
plt.show()
```
